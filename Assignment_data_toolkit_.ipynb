```python
# Python Program illustrating
# numpy.sum() method
import numpy as np

# 2D array
arr = [[14, 17, 12, 33, 44],
    [15, 6, 27, 8, 19],
    [23, 2, 54, 1, 4,]]

print("\nSum of arr : ", np.sum(arr))
print("Sum of arr(axis = 0) : ", np.sum(arr, axis = 0))
print("Sum of arr(axis = 1) : ", np.sum(arr, axis = 1))

print("\nSum of arr (keepdimension is True): \n",
    np.sum(arr, axis = 1, keepdims = True))
```

```python
import pandas as pd
data = {'Name': ['Bob', 'Charlie', 'David'],
        'Age': [23, 45, 28],
        'Score': [90, 78, 88]}
df = pd.DataFrame(data)

# AND operation: Age > 25 AND Score > 80
and_filter = df[(df['Age'] > 25) & (df['Score'] > 80)]
print("AND Operation Result:")
print(and_filter)

# OR operation: Age > 25 OR Score > 80
or_filter = df[(df['Age'] > 25) | (df['Score'] > 80)]
print("\nOR Operation Result:")
print(or_filter)

# NOT operation: NOT (Age > 25)
not_filter = df[~(df['Age'] > 25)]
print("\nNOT Operation Result:")
print(not_filter)
```
